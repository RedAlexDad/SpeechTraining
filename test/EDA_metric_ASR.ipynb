### Промежуточный вывод

Как и видим, что в большинствах датасетах лидирует модель `wav2vec2-mbart50-ru`, но нельзя заранее опеределить, пока не получим медианное значение по всем метрикам, сгруппируя по названию моделей

### Группировка по названию модели

```python
df_grouped = df_full.groupby(['MODEL']).agg({
    'WER': 'median',
    'CER': 'median',
    'MER': 'median',
    'WIL': 'median',
    'SAMPLES': 'first'
}).reset_index()

display(df_grouped)

# Сохраним объединенных датафреймов
df_grouped.to_excel('df_grouped_metric.xlsx', index=False)
```

#### Установка рейтингов по лучшей модели

\begin{equation}
Рейтинг = \frac{{RATING_{max} - RATING}}{{RATING_{max} - RATING_{min}}} \times 100
\end{equation}

```python
# Инвертирование рейтинга (чем меньше, тем лучше)
df_grouped['RATING'] = df_grouped[['WER', 'CER', 'MER', 'WIL']].sum(axis=1)
```

```python
# Нормализация рейтинга для получения 100% соотношения
df_grouped['RATING'] = (df_grouped['RATING'].max() - df_grouped['RATING']) / (df_grouped['RATING'].max() - df_grouped['RATING'].min()) * 100
```

```python
# Сортировка по рейтингу
df_grouped = df_grouped.sort_values(by='RATING', ascending=False)
```

```python
# Добавим номера к названиям моделей для упрощения отображения на графике
df_grouped['MODEL_SHORT'] = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5']
```

```python
# Обновление индексов
df_grouped.reset_index(drop=True, inplace=True)
```

```python
# Сохраним объединенных датафреймов
df_grouped.to_excel('df_grouped_metric_with_rating.xlsx', index=False)
```

#### График

```python
# Сортировка по рейтингу
df_grouped = df_grouped.sort_values(by='RATING', ascending=True)

plt.figure(figsize=(10, 6))

bars = plt.barh(df_grouped['MODEL'], df_grouped['RATING'])
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.title('Рейтинг моделей')
plt.xlabel('Рейтинг (%)')
plt.ylabel('Модель')

# Добавление значений рейтинга в конце каждой полоски
for bar, rating in zip(bars, df_grouped['RATING']):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2, f'{rating:.2f}%', va='center')

plt.show()
```

```python
# Проверим индексы и отсортируем данные по MODEL_SHORT
df_grouped = df_grouped.sort_values(by='MODEL_SHORT').reset_index(drop=True)

titles = ['Сравнение WER', 'Сравнение CER', 'Сравнение MER', 'Сравнение WIL']

# Создание графиков
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for i, ax in enumerate(axs.flat):
    ax.plot(df_grouped['MODEL_SHORT'], df_grouped[metrics[i]], marker='o')
    
    # Добавление значений к каждой точке
    for j, txt in enumerate(df_grouped[metrics[i]]):
        ax.text(j, df_grouped[metrics[i]][j] + 0.01, f'{txt:.3f}', ha='center')
    
    ax.set_title(titles[i])
    ax.set_xlabel('Model')
    ax.set_ylabel(metrics[i])
    ax.grid(True)

plt.tight_layout()
plt.show()
```
